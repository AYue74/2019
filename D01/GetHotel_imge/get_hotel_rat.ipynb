```python
##### # coding: utf-8

# In[ ]:


from selenium import webdriver
from bs4 import BeautifulSoup

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import time

hotel_url = []
hotel_name = []
hotel_id = []

with open('givehotel','r',encoding='utf-8') as f:
    for line in f:
        a = line.split()
        hotel_name.append(a[2])
        hotel_url.append(a[-1])
        hotel_id.append(a[0])


chrome_options = webdriver.ChromeOptions()
chrome_options.set_headless() # 设置启动无界面化 (沒有瀏覽器視窗)
prefs = {"profile.managed_default_content_settings.images": 2}#設置禁止圖片加載
chrome_options.add_experimental_option("prefs", prefs)
# browser = webdriver.Chrome('C:/Users/asmin/Desktop/cc/chromedriver.exe'                           ,chrome_options=chrome_options)
browser = webdriver.Chrome('C:/Users/asmin/Desktop/cc/GetHotel_imge/chromedriver.exe'
                          ,options = chrome_options)# 就不會有警告提示 

fp = open('hotel_rat','a+',encoding='utf-8')
j = -1

while j<6201:
    try:

        for i in range(j+1,len(hotel_url)):
            print(i)
            j = i
            browser.get(hotel_url[i])
            try:
                browser.find_element_by_xpath('//*[@id="taplc_trip_planner_breadcrumbs_0"]').click()
            except:
                pass
            locator = (By.CSS_SELECTOR, "span[class='hotels-hotel-review-about-with-photos-Reviews__overallRating--vElGA']")
            WebDriverWait(browser, 3, 0.2).until(EC.presence_of_element_located(locator))
            rating = browser.find_element_by_class_name('hotels-hotel-review-about-with-photos-Reviews__overallRating--vElGA')
            hotel_rat = rating.text
            print(hotel_name[i],hotel_rat)
            text = hotel_name[i]+'   '+hotel_rat+'\n'
            fp.write(text)
    except:
        continue
    
        
    
fp.close()
```
